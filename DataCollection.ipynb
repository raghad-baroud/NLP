```python
def Expo_search_query(query):
  options={'since' : "2021-10-1",
          'until' : "2022-10-1" }
  query=twikit.utils.build_query(query,options)
  return query
```

```python
def retrieve_tweets(query,count=50):
  try:
    tweets = client.search_tweet(query, 'Top',count=count)
  except IndexError:
    return pd.DataFrame() #return empty dataset for the 'list out of bound' error

  all_tweets=[]
  while len(tweets)!=0:
    all_tweets+=list(tweets)
    try:
      tweets=tweets.next()
    except :
      break


  tweets_dataset=[]
  for tweet in all_tweets:

    sample=[tweet.id,tweet.text]
    tweets_dataset.append(sample)

  return pd.DataFrame(tweets_dataset,columns=['Tweet_ID','Tweet_Text'])
```

#### Quick Data Cleaning

To remove the non-necessary tweets, we will take the following actions:

- Remove duplicates tweets
- Remove Tweets with the exact same `Tweet_Text`
- Remove non-Arabic tweets
- Remove Tweets already exist in the pre-collected dataset `tweets_classified`

```python
d=final_data.drop_duplicates() #duplicates rows
len(d)
```

```python
d=d.drop_duplicates(subset=['Tweet_Text']) #duplicates texts
len(d)
```

```python
!pip install pyarabic
clear_output()
import pyarabic.trans


indecies=[]
for i,smp in enumerate(d.values):
  seg=pyarabic.trans.segment_language(smp[1])
  no_arabic=True
  for s in seg:      #detect non-arabic tweets
    if s[0]=='arabic':
      no_arabic=False
      break
  if no_arabic:
    indecies.append(i)

d.iloc[indecies]
```

```python
d=d.drop(indecies,axis=0) #remove non-arabic
len(d)
```

```python
d['DataSource']=1 #first dataset
d.head()
```

```python
precollected_data=pd.read_excel('/content/tweets_classified.xlsx')
precollected_data.drop(columns=['Column1', 'index', 'Timestamp', 'Likes', 'Retweets','Length', 'Date', 'Time', 'Class'],inplace=True)
precollected_data.columns=['Tweet_ID','Tweet_Text']
precollected_data['DataSource']=2 #second dataset
precollected_data.head()
```

```python
full_dataset=pd.concat([precollected_data,d])
print('Before',len(full_dataset))
full_dataset.drop_duplicates(subset=['Tweet_ID'],inplace=True) #drop using the tweet id
print('After', len(full_dataset))
```

```python
# return back our collected data
d=full_dataset[full_dataset['DataSource']==1]
d.drop(columns=['DataSource'],inplace=True)
len(d)
```
